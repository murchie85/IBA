```python
# ======================== Python dependencies
import json
import pandas
import datetime
import os
import statistics
import numpy as np



```

```python

# ************************************************************************
#                                  CLASS SECTION                         *
# ************************************************************************
# PROGRAM NAME: Prototype                                                *
# DESCRIPTION: This is a proving program to test transaction processing  *
#              capabilities and process csv files                        *
#                                                                        *
#*************************************************************************
# Amendments:                                                            *
# Name:                                                                  *
# Date:                                                                  * 
#*************************************************************************








# ======================== collate filenames to array

class processFiles:
    def __init__(self, directory): # initialize   
        self.directory = directory
    
    def returnAvailableAccounts(self):
        availableAccounts = []
        for filename in os.listdir(self.directory):
            if filename.endswith(".csv"): 
                #print(os.path.join(self.directory, filename))
                AccountFile = filename[:-4]
                availableAccounts.append(AccountFile)
                continue
            else:
                continue
        return availableAccounts


class accountToMatrix:
    def __init__(self, availableAccounts):
        self.availableAccounts = availableAccounts
    
    def generateMatrix(self):
        a = 0
        currentRow = 0
        accountArray = []
        for a in range(0, len(self.availableAccounts)):
            print("The current Account Being processed is " + str(availableAccounts[a]))
            with open(str(directory) + str(self.availableAccounts[a]) + ".csv", 'rU') as f:
                currentRow = []
                for line in f:
                    words = line.split(",")
                    myarray = np.asarray(words)
                    
                    #clear empty rows and title row
                    if len(myarray[0]) != 10:
                        continue

                    print(str(myarray))
                    currentRow.append(myarray)
                accountArray.append(np.asarray(currentRow))
        return accountArray




directory = "../../IBA-DATA/data/"


# construct object from processFiles class
fileObj = processFiles(directory)
availableAccounts = fileObj.returnAvailableAccounts()

print('processing files ...................')
print('The available files are ' + str(availableAccounts))



accountArrayOBJ = accountToMatrix(availableAccounts)
accountArray = accountArrayOBJ.generateMatrix()


```
